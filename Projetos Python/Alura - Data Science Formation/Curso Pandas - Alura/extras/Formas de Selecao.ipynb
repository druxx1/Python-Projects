```python
import pandas as pd
```

```python
data = [(1, 2, 3, 4),
        (5, 6, 7, 8),
        (8, 10, 11, 12),
        (13, 14, 15, 16)]
df = pd.DataFrame(data = data, index = "l1 l2 l3 l4".split(), columns = "c1 c2 c3 c4".split())
df
```

```python
df['c1']
```

```python
df[['c3','c1']]
```

```python
# O ':' seleciona todas as linhas e colunas

df[:]
```

```python
# A l1 eh a linha 0 no indice, por isso seleciona a l2

# A l4(linha 3 no indice) nao eh selecionada, e sim a l3(linha 2 no indice) pois a selecao nao chega ate o ultimo numero
# do indice ([1:3] = l2, l3,)

df[1:3]
```

```python
df[1:][['c3', 'c1']]
```

```python
df.loc['l3']
```

```python
df.loc[['l3', 'l2']]
```

```python
df.loc['l1', 'c2']
```

```python
# loc a partir dos indices (iloc = index loc)

df.iloc[0, 1]
```

```python
df.loc[['l3', 'l1'], ['c4', 'c1']]
```

```python
df.iloc[[2, 0], [3, 0]]
```

## DataFrame de referencia:

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
alunos
```

```python
# Crie um DataFrame somente com os alunos reprovados e mantenha neste DataFrame apenas as colunas Nome, Sexo e Idade,
# nesta ordem.
slc = alunos.Aprovado == False
alunos.loc[slc, ['Nome', 'Sexo', 'Idade']]
# ou alunos[['Nome', 'Sexo', 'Idade']][slc]
```

```python
# Crie uma visualização com os três alunos mais novos.
alunos.sort_values(by = 'Idade', inplace = True)
alunos.iloc[:3]
# ou alunos[:3]
```
