```python
fig = plt.figure(figsize = (15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.7,0.65,0.3,0.31])

eixo.grid(True)
eixo.plot(dados['data'], dados['temperatura'], color = 'red')
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.set_ylim(270,320)
eixo.set_title("Temperatura em Maio de 2014", fontsize = 25, pad = 20)
eixo.legend(['temperatura'], loc = 'lower left', fontsize = 15)
eixo.set_ylabel("Temperatura", fontsize = 15)
eixo.set_xlabel("Data", fontsize = 15)

azul_esquerda = dados['data'] < datetime.datetime(2014,5,1)
azul_direita = dados['data'] > datetime.datetime(2014,6,1)

eixo2.grid(True)
eixo2.plot(dados['data'], dados['temperatura'], color = 'red')
eixo2.plot(dados[azul_esquerda]['data'], dados[azul_esquerda]['temperatura'], color = 'darkblue', lw = 1.75)
eixo2.plot(dados[azul_direita]['data'], dados[azul_direita]['temperatura'], color = 'darkblue', lw = 1.75)
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo2.set_title("Temperatura em 2014", fontsize = 17)
eixo2.legend(['temperatura'], loc = 'upper left', fontsize = 10)
eixo2.set_ylabel("Temperatura", fontsize = 12)
eixo2.set_xlabel("Data", fontsize = 12)
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = 'green')
eixo.set_title("Temperatura em Maio de 2014", fontsize = 20, pad = 15)
eixo.set_xlabel("Data", fontsize = 13)
eixo.set_ylabel("Temperatura", fontsize = 13)
eixo.legend(["temperatura"], loc = 'lower right', fontsize = 13)
eixo.grid(True)

eixo.axhline(max(dados['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color = 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = 'green')
eixo.set_title("Temperatura em Maio de 2014", fontsize = 20, pad = 15)
eixo.set_xlabel("Data", fontsize = 13)
eixo.set_ylabel("Temperatura", fontsize = 13)
eixo.legend(["temperatura"], loc = 'lower right', fontsize = 13)
eixo.grid(True)


x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])
eixo.annotate("Max", xy = (x1,y1), fontsize = 15)

x2 = dados['data'][dados['temperatura'].idxmin()]
y2 = min(dados['temperatura'])
eixo.annotate("Min", xy = (x2,y2), fontsize = 15)


eixo.axhline(max(dados['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color = 'k', linestyle = '--')
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados['data'], dados['temperatura'], color = 'green')
eixo.set_title("Temperatura em Maio de 2014", fontsize = 20, pad = 15)
eixo.set_xlabel("Data", fontsize = 13)
eixo.set_ylabel("Temperatura", fontsize = 13)
eixo.legend(["temperatura"], loc = 'lower right', fontsize = 13)
eixo.grid(True)


x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])
x2 = dados['data'][dados['temperatura'].idxmax() + 2500]
y2 = max(dados['temperatura']) - 5
eixo.annotate("Max", xy = (x1,y1), fontsize = 15, xytext = (x2,y2), arrowprops = dict(facecolor = 'black'))

x3 = dados['data'][dados['temperatura'].idxmin()]
y3 = min(dados['temperatura'])
x4 = dados['data'][dados['temperatura'].idxmin() - 4250]
y4 = min(dados['temperatura']) + 4
eixo.annotate("Min", xy = (x3,y3), fontsize = 15, xytext = (x4,y4), arrowprops = dict(facecolor = 'black'))


eixo.axhline(max(dados['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color = 'k', linestyle = '--')
```

```python
temperatura_por_dia_da_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize = (7,5))
eixo = fig.add_axes([0,0,1,1])
cores = ['green', 'darkmagenta', 'darkblue', 'lightblue', 'gold', 'orange', 'red']

eixo.bar(nome_dias, temperatura_por_dia_da_semana, color = cores, edgecolor = 'black')
eixo.set_title("Temperatura por dia da semana", fontsize = 20)
eixo.set_ylabel("Temperatura Media", fontsize = 13)
eixo.set_xlabel("Dias", fontsize = 13)
```

```python
fig = plt.figure(figsize = (5,5))
eixo = fig.add_axes([0,0,1,1])
finais_de_semana = [0.1, 0, 0, 0, 0, 0.1, 0.1]

eixo.pie(temperatura_por_dia_da_semana, labels = nome_dias, autopct = '%.3f%%', 
         explode = finais_de_semana, shadow = True)
eixo.set_title("Temperatura por Dia da Semana", fontsize = 20)
```
