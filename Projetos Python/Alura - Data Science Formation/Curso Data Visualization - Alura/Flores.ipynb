```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dados = pd.read_csv('dados/iris.csv')
```

```python
dados.head(10)
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

eixo.scatter(x = dados['comprimento_sépala'], y = dados['largura_sépala'])
eixo.set_title("Grafico de Dispersao", fontsize = 19)
eixo.set_xlabel("Comprimento da Sepala", fontsize = 13)
eixo.set_ylabel("Largura da Sepala", fontsize = 13)
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor': 'darkblue', 'Iris-virginica': 'green'}

for especie in dados['espécie'].unique():
    temporario = dados[dados['espécie'] == especie]
    eixo.scatter(temporario['comprimento_sépala'], temporario['largura_sépala'], color = cores[especie])
    
eixo.set_title("Grafico de Dispersao", fontsize = 23)
eixo.set_xlabel("Comprimento da Sepala", fontsize = 15)
eixo.set_ylabel("Largura da Sepala", fontsize = 15)
eixo.tick_params(labelsize = 15)
eixo.legend(cores, fontsize = 15)
```

```python
fig = plt.figure(figsize = (11,6))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa': 'red', 'Iris-versicolor': 'darkblue', 'Iris-virginica': 'green'}
marcadores = {'Iris-setosa': 'x', 'Iris-versicolor': 'o', 'Iris-virginica': 'v'}

for especie in dados['espécie'].unique():
    temporario = dados[dados['espécie'] == especie]
    eixo.scatter(temporario['comprimento_sépala'], temporario['largura_sépala'], color = cores[especie],
                 marker = marcadores[especie], s = 100)
    
eixo.set_title("Grafico de Dispersao", fontsize = 23, pad = 10)
eixo.set_xlabel("Comprimento da Sepala", fontsize = 15)
eixo.set_ylabel("Largura da Sepala", fontsize = 15)
eixo.tick_params(labelsize = 15)
eixo.legend(cores, fontsize = 15)
```

```python
fig = plt.figure(figsize = (7,5))
eixo = fig.add_axes([0,0,1,1])

eixo.boxplot(dados['largura_pétala'])
eixo.set_title("Grafico de Caixa", fontsize = 23, pad = 10)
eixo.set_xticklabels(['largura_pétala'])
```

```python
fig = plt.figure(figsize = (7,5))
eixo = fig.add_axes([0,0,1,1])

cores = ['pink', 'mediumpurple', 'darkcyan', 'lightcyan']

caixas = eixo.boxplot(dados.drop('espécie', axis = 1), patch_artist = True)
eixo.set_title("Grafico de Caixa", fontsize = 20)
eixo.set_xticklabels(dados.drop('espécie', axis = 1))

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color = cor)
```

```python
fig = plt.figure(figsize = (7,5))
eixo = fig.add_axes([0,0,1,1])

cores = ['pink', 'mediumpurple', 'darkcyan', 'lightcyan']

caixas = eixo.boxplot(dados.drop('espécie', axis = 1), patch_artist = True)
eixo.set_title("Grafico de Caixa", fontsize = 20)
eixo.set_xticklabels(dados.drop('espécie', axis = 1))

for caixa, cor in zip(caixas['boxes'], cores):
    caixa.set(color = cor)
    
for outlier in caixas['fliers']:
    outlier.set(marker = 'x', markersize = 5)
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(dados['comprimento_pétala'], bins = 20)
eixo.set_title("Histograma", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(dados['comprimento_pétala'], bins = 20, density = True)
eixo.set_title("Histograma", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = dados['comprimento_pétala'].mean(), dados['comprimento_pétala'].std()

eixo.hist(dados['comprimento_pétala'], bins = 20)
eixo.set_title("Histograma", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma), xy = (6.1, 31), fontsize = 20, alpha = 0.85)

eixo.axvline(mu, color = 'red', linestyle = '--', alpha = 0.6)
eixo.annotate("Media", xy = (mu - 0.55, 25.3), fontsize = 13, color = 'red')

eixo.axvline(dados['comprimento_pétala'].median(), color = 'green', linestyle = '--', alpha = 1)
eixo.annotate("Mediana", xy = (dados['comprimento_pétala'].median() + 0.05, 25.3), fontsize = 13, color = 'green')
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

iris_vc = dados[dados['espécie'] == 'Iris-versicolor']
mu, sigma = iris_vc['comprimento_pétala'].mean(), iris_vc['comprimento_pétala'].std()

eixo.hist(iris_vc['comprimento_pétala'], bins = 20)
eixo.set_title("Histograma Iris-versivcolor", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma), xy = (3.06, 6.15), fontsize = 20, alpha = 0.85)

eixo.axvline(mu, color = 'red', linestyle = '--', alpha = 0.6)
eixo.annotate("Media", xy = (mu - 0.2, 7.07), fontsize = 13, color = 'red')

eixo.axvline(iris_vc['comprimento_pétala'].median(), color = 'green', linestyle = '--', alpha = 1)
eixo.annotate("Mediana", xy = (iris_vc['comprimento_pétala'].median() + 0.02, 7.07), fontsize = 13, color = 'green')

fig.savefig('histograma_irisvc.png', bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

iris_st = dados[dados['espécie'] == 'Iris-setosa']
mu, sigma = iris_st['comprimento_pétala'].mean(), iris_st['comprimento_pétala'].std()

eixo.hist(iris_st['comprimento_pétala'], bins = 20)
eixo.set_title("Histograma Iris-setosa", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma), xy = (1.02, 12.3), fontsize = 20, alpha = 0.85)

eixo.axvline(mu, color = 'red', linestyle = '--', alpha = 0.6)
eixo.annotate("Media", xy = (mu - 0.088, 14.2), fontsize = 13, color = 'red')

eixo.axvline(iris_st['comprimento_pétala'].median(), color = 'green', linestyle = '--', alpha = 1)
eixo.annotate("Mediana", xy = (iris_st['comprimento_pétala'].median() + 0.01, 14.2), fontsize = 13, color = 'green')

fig.savefig('histograma_irisst.png', bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize = (7,6))
eixo = fig.add_axes([0,0,1,1])

iris_vg = dados[dados['espécie'] == 'Iris-virginica']
mu, sigma = iris_vg['comprimento_pétala'].mean(), iris_vg['comprimento_pétala'].std()

eixo.hist(iris_vg['comprimento_pétala'], bins = 20)
eixo.set_title("Histograma Iris-virginica", fontsize = 23)
eixo.set_xlabel("Comprimento da Pétala", fontsize = 13)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma), xy = (4.54, 6.6), fontsize = 20, alpha = 0.85)

eixo.axvline(mu, color = 'red', linestyle = '-', alpha = 0.6)
eixo.annotate("Media", xy = (mu + 0.02, 7.3), fontsize = 13, color = 'red')

eixo.axvline(iris_vg['comprimento_pétala'].median(), color = 'green', linestyle = '--', alpha = 0.8)
eixo.annotate("Mediana", xy = (iris_vg['comprimento_pétala'].median() + 0.02, 7), fontsize = 13, color = 'green')

fig.savefig('histograma_irisvg.png', bbox_inches = 'tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_irisvc.png').size
combinada = Image.new('RGB', (3 * largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_irisvc.png', 'histograma_irisst.png', 'histograma_irisvg.png']):
    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura
    
combinada.save('combinada.png')
```
