```python
#  features (0= nao, 1= sim)
# pelo longo? -> sim ou nao
# perna curta? -> sim ou nao
# faz auau? -> sim ou nao
porco1= [0, 1, 0]
porco2= [0, 1, 1]
porco3= [1, 1, 0]

cachorro1= [0, 1, 1]
cachorro2= [1, 0, 1]
cachorro3= [1, 1, 1]

# 0-> cachorro & 1-> porco
dados= [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
classes= [1,1,1,0,0,0]
```

### 'Dados' eh um nome muito generico, pois meio que qualquer coisa pode ser um dado (nome do modelo de celular eh um dado, meu nome eh um dado, 6 animais que vao ser classificados por um algoritimo sao dados, assim vai...).
### Em geral, esses 'dados' que sao o conjunto de features que a gente vai analisar, a gente pensa neles como em uma funcao matematica ( f(x) = y ), entao o modelo funciona como uma funcao matematica, onde os 'dados' sao x e as 'classes' sao y.
### x -> dados com as features & y -> classes, que as vezes tambem sao chamadas de labels.
### Porem esses dados que estamos utilizando sao para treinar o nosso algoritimo, entao o nome que a gente usa de verdade eh o x de treino e y de treino.

```python
train_x= [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
train_y= [1,1,1,0,0,0]
```

```python
from sklearn.svm import LinearSVC

model= LinearSVC()

# model.fit(dados, classes) -> Dada a explicacao da celula acima, nao usamos mais 'dados' e 'classes' como variaveis e sim,
# x e y, entao vai ficar:
model.fit(train_x, train_y)
```

```python
animal_misterioso= [1, 1, 1]
model.predict([animal_misterioso])
```

```python
misterio1= [1, 1, 1]
misterio2= [1, 1, 0]
misterio3= [0, 1, 1]

# teste= [misterio1, misterio2, misterio3] -> Dada a explicao da primeira celula, a variavel 'teste' tambem se enquadra  
# como x, pois tambem sao dados, porem agora sao os dados para testar, ou seja, nao sera 'train_x' e sim 'test_x'.
test_x= [misterio1, misterio2, misterio3]

# valor_original_teste= [0, 1, 1] -> A mesma coisa se aplica a variavel 'valor_original_teste', ja que ela se encaixa como
# a varivel que classificara o teste, ou seja, sera a label do teste. Portanto a variavel vai se chamar 'test_y'.
test_y= [0, 1, 1]
```

```python
previsoes= model.predict(test_x)
print(previsoes)

#previsoes == valor_original_teste
previsoes == test_y
```

```python
#corretos= (previsoes == valor_original_teste).sum()
corretos= (previsoes == test_y).sum()
#total= len(teste)
total= len(test_x)
taxa_acerto= corretos / total

print("Taxa de acerto: {:.2f}%".format(taxa_acerto*100))
```

```python
from sklearn.metrics import accuracy_score

#taxa_acerto= accuracy_score(valor_original_teste, previsoes)
taxa_acerto= accuracy_score(test_y, previsoes)
print("Taxa de acerto: {:.2f}%".format(taxa_acerto*100))
```
