```python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:,-1] # iloc[:,-1] seleciona todas as linhas, porem somente a 
                                                                  # ultima coluna do DataFrame(realizara a mesma funcao
                                                                  # da variavel 'last_date' criada contendo a ultima data)

manipulated_recovered=round((latest_by_country(confirmed)-latest_by_country(deaths))/2).astype(int)
manipulated_recovered
```

```python
informations = [latest_by_country(confirmed), latest_by_country(deaths), manipulated_recovered]
concated = pd.concat(informations, axis = 1)
concated.columns = ["Confirmed", "Deaths", "Recovered"]
concated.index.names = [""]
concated.columns.names = ["Country/Region"]
concated.drop(index = ['Korea, North'], inplace = True)
concated
```

```python
# taxa de letalidade eh expressa em %
# o numero de casos letais dentro dos casos que foram letais e nao letais

# imagina que existe uma doenca que em janeiro de 2020 afetou 1000 pessoas
# 50 pessoas morreram (por exemplo), ou seja, a letalidade eh de 5%
# e a doenca deixou de existir para sempre
# caso a doenca continue existindo a taxa de letalidade podera ser outra
# tanto para cima (+ de 5%) quanto para baixo (- de 5%)

# eh muito dificil falar de letalidade tentando ser preciso durante uma
# epidemia, pois estao aparecendo casos, fazendo com que a taxa varie

# no comeco eh comum que apenas casos graves sejam reportados
```

```python
sum_up = concated.sum()
letality_rate_1 = sum_up['Deaths'] / sum_up['Confirmed']
letality_rate_2 = sum_up['Deaths'] / (sum_up['Recovered'] + sum_up['Deaths'])
print('Letality Rate 1: {0:.2f}%'.format(letality_rate_1 * 100),
      '\nLetality Rate 2: {:.2f}%'.format(letality_rate_2 * 100))
```

```python
def latest_by_country_at (data, date):
    return data.groupby('Country/Region').sum()[date]


informations_2_21_20 = [latest_by_country_at(confirmed, '2/21/20'), latest_by_country_at(deaths, '2/21/20'),
                       latest_by_country_at(recovered, '2/21/20')]
concated_2_21_20= pd.concat(informations, axis = 1)
concated_2_21_20.columns = ["Confirmed", "Deaths", "Recovered"]


sum_up = concated_2_21_20.sum()
letality_rate_1 = sum_up['Deaths'] / sum_up['Confirmed']
letality_rate_2 = sum_up['Deaths'] / (sum_up['Recovered'] + sum_up['Deaths'])
print('Letality Rate 1: {0:}%'.format(letality_rate_1 * 100),
      '\nLetality Rate 2: {:}%'.format(letality_rate_2 * 100))
```

```python
letality_rate_1 = (concated['Deaths'] / concated['Confirmed']) * 100
letality_rate_2 = concated['Deaths'] / (concated['Recovered'] + concated['Deaths']) * 100

concated['Letality Rate 1'] = letality_rate_1.round(2)
concated['Letality Rate 2'] = letality_rate_2.round(2)

concated.head()
```

```python
# vou assumir algumas premissas
# estou assumindo que na UN China = John Hopkins Mainland China

def rename_location (location):
    if location == 'China':
        return 'China' # era para retornar 'Mainland China', porem durante o curso assumi que 'Mainland China' = 'China'
    if location == 'Iran (Islamic Republic of)':
        return 'Iran'
    if location == 'Russian Federation':
        return 'Russia'
    if location == 'Republic of Korea':
        return 'Korea, South'
    if location == 'United States of America':
        return 'US'
    else:
        return location
        

un_population_total['Location for WHO'] = un_population_total.index.map(rename_location)
un_population_total.query('`Location for WHO` == "South Korea"')
```

```python
un_population_total = un_population_total.set_index('Location for WHO')
un_population_total.query('index == "Russia"')
```

```python
grouped_per_day = pd.DataFrame([confirmed.sum(numeric_only = True)[2:-1], recovered.sum(numeric_only = True)[2:],
                                deaths.sum(numeric_only = True)[2:]], index = ['Confirmed', 'Recovered', 'Deaths']).T
grouped_per_day.head(10)
```

```python
grouped_per_day['Letality Rate 1'] = grouped_per_day['Deaths'] / grouped_per_day['Confirmed'] * 100
grouped_per_day['Letality Rate 2'] = grouped_per_day['Deaths'] / (grouped_per_day['Deaths'] + 
                                                                  grouped_per_day['Recovered']) * 100
grouped_per_day
```

```python
grouped_per_day['Letality Rate 1'][:'3/3/20'].plot()
plt.title("Mortes por Casos Confirmados no Mundo Todo", pad = 10, fontsize = 15)
plt.show()
```

```python
grouped_per_day['Letality Rate 2'][:'3/3/20'].plot()
plt.title("Casos por Recuperados ou Mortos no Mundo Todo", pad = 10, fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (10,5))
grouped_per_day['Letality Rate 1'][:'3/3/20'].plot()
ax = grouped_per_day['Letality Rate 2'][:'3/3/20'].plot()
plt.legend(['Mortes / Casos Confirmados', 'Mortes / (Mortes + Recuperados)'])
plt.title("Letalidade 'Pessimista e Otimista' no Mundo Todo, Convergindo para o Mesmo Valor", pad = 15, fontsize = 15)
plt.grid(linestyle = '-')
ax.set_xlabel("Fonte\nDados: Johns Hopkins - https://github.com/CSSEGISandData/COVID-19\nCurso: Alura - https://www.alura.com.br",
             labelpad = 10)
ax.set_ylabel("% de Letalidade", labelpad = 10)
plt.show()
```
