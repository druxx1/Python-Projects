Consider that you have been hired by a Multiplex management organization to extract the information of the top 50 movies with the best average rating from the web link shared below.
<br><br>
https://web.archive.org/web/20230902185655/https://en.everybodywiki.com/100_Most_Highly-Ranked_Films
<br><br>
The information required is ***Average Rank, Film, and Year***.
You are required to write a Python script that extracts the information and saves it to a CSV file ***top_50_films.csv***. You are also required to save the same information to a database ***Movies.db*** under the table name ***Top_50***.

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import sqlite3
```

```python
url = "https://web.archive.org/web/20230902185655/https://en.everybodywiki.com/100_Most_Highly-Ranked_Films"
db_name = "Movies.db"
table_name = "Top_50"
csv_name = "top_50_films.csv"
df = pd.DataFrame(columns = ["Average Rank", "Film", "Year"])
count = 0
```

# Creating the CSV file

```python
html_page = requests.get(url).text
data = BeautifulSoup(html_page, 'html.parser')
data
```

```python
tables = data.find_all('tbody')
rows = tables[0].find_all('tr')
```

```python
for row in rows:
    if count < 50:
        data_obj = row.find_all('td')
        # print(data_obj)
        if len(data_obj) != 0:
            data_dict = {"Average Rank": data_obj[0].contents[0],
                             "Film": data_obj[1].contents[0],
                             "Year": data_obj[2].contents[0]}
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df,df1], ignore_index=True)
            count+=1
    else:
        break

df
```

```python
df.to_csv(csv_name)
```

# Creating the SQL file

```python
conn = sqlite3.connect(db_name)
df.to_sql(table_name, conn, if_exists= 'replace', index= False)
conn.close()
```
