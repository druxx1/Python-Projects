Consider a dataset of employee records that is available with an HR team in a CSV file. As a Data Engineer, you are required to create the database called STAFF and load the contents of the CSV file as a table called INSTRUCTORS. The headers of the available data are :<br><br>

<br>
1. Create a database using Python

2. Load the data from a CSV file as a table to the database

3. Run basic "queries" on the database to access the information

```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("STAFF.db")
```

```python
attributes = ['ID', 'FNAME', 'LNAME', 'CITY', 'CCODE']
df = pd.read_csv("INSTRUCTOR.csv", names = attributes)
df
```

```python
# if_exists = 'fail' -> Default. The command doesn't work if a table with the same name exists in the database.
# if_exists = 'replace' -> The command replaces the existing table in the database with the same name.
# if_exists = 'append' -> The command appends the new data to the existing table with the same name.

table_name = "INSTRUCTOR"
df.to_sql(table_name, conn, if_exists = 'replace', index =False)
print('Table is ready')
```

```python
query = f"SELECT * FROM {table_name}"
print(query)
pd.read_sql(query, conn)
```

```python
query = f"SELECT FNAME FROM {table_name}"
print(query)
pd.read_sql(query, conn)
```

```python
query = f"SELECT COUNT(*) FROM {table_name}"
print(query)
pd.read_sql(query, conn)
```

```python
data_dict = {'ID' : [100],
            'FNAME' : ['John'],
            'LNAME' : ['Doe'],
            'CITY' : ['Paris'],
            'CCODE' : ['FR']}

data_append = pd.DataFrame(data_dict)

data_append.to_sql(table_name, conn, if_exists= 'append', index= False)
print('Data appended successfully')
```

```python
pd.read_sql(F"SELECT * FROM {table_name}", conn)
```

```python
conn.close()
```
