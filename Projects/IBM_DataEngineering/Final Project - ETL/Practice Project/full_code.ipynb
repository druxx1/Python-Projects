# Project Scenario
An international firm that is looking to expand its business in different countries across the world has recruited you. You have been hired as a junior Data Engineer and are tasked with creating an automated script that can extract the list of all countries in order of their GDPs in billion USDs (rounded to 2 decimal places), as logged by the International Monetary Fund (IMF). Since IMF releases this evaluation twice a year, this code will be used by the organization to extract the information as it is updated.

You can find the required data on this webpage: https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29

The required information needs to be made accessible as a CSV file **Countries_by_GDP.csv** as well as a table **Countries_by_GDP** in a database file **World_Economies.db** with attributes **Country** and **GDP_USD_billion.**

Your boss wants you to demonstrate the success of this code by running a query on the database table to display only the entries with more than a 100 billion USD economy. Also, log the entire process of execution in a file named **etl_project_log.txt.**

You must create a Python code that performs all the required tasks.<br><br>

# Objectives
1. Write a data extraction function to retrieve the relevant information from the required URL.

2. Transform the available GDP information into *Billion USD* from *Million USD*.

3. Load the transformed information to the required CSV file and as a database file.

4. Run the required query on the database.

5. Log the progress of the code with appropriate timestamps.



## Preliminary

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import sqlite3 as sql
from datetime import datetime
```

```python
url = 'https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29'
table_attr = ["Country", "GDP_USD_millions"]
db_name = 'World_Economies.db'
table_name = 'Countries_by_GDP'
csv_path = 'Countries_by_GDP.csv'
```

## Code

```python
# 1. extract
def extract(url, table_attr):
    page = requests.get(url).text
    data = BeautifulSoup(page, 'html.parser')
    df = pd.DataFrame(columns= table_attr)
    tables = data.find_all('tbody')
    rows = tables[2].find_all('tr')
    for row in rows:
        col = row.find_all('td')
        if len(col) != 0:
            if col[0].find('a') is not None and '—' not in col[2]:
                ddict = {
                    "Country": col[0].a.contents[0],
                    "GDP_USD_millions": col[2].contents[0]
                }
                df1 = pd.DataFrame(ddict, index= [0])
                df = pd.concat([df,df1], ignore_index= True)
    return df
```

```python
df = extract(url, table_attr)
df
```

```python
# 2. transform
def transform(df):
    gdp_list = df['GDP_USD_millions'].tolist()
    updated_gdp = list()
    gdp_billion = list()

    for item in gdp_list:
        updated_gdp.append(float(item.replace(",","")))
    for nmb in updated_gdp:
        gdp_billion.append(round(nmb/1000, 2))
        
    df["GDP_USD_millions"] = gdp_billion
    df = df.rename(columns= {"GDP_USD_millions":"GDP_USD_billions"})
    
    return df
```

```python
df = transform(df)
df
```

```python
# 3. load
def load_to_csv(df, csv_path):
    df.to_csv(csv_path)

def load_to_db(df, sql_conn, table_name):
    df.to_sql(table_name, sql_conn, if_exists= 'replace', index= False)
```

```python
# 4. run query
def run_query(query_stmt, sql_conn):
    print(query_stmt)
    output = pd.read_sql(query_stmt, sql_conn)
    print(output)
```

```python
# 5. log
def log_process(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Year-Monthname-Day-Hour-Minute-Second 
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("etl_project_log.txt", "a") as file:
        file.write(timestamp + ": " + message + "\n")
```

## Running the project

```python
log_process("Preliminaries complete. Initiating ETL process.")

df = extract(url, table_attr)

log_process("Data extraction complete. Initiating Transformation process")

df = transform(df)

log_process("Data transformation complete. Initiating loading process.")

load_to_csv(df, csv_path)

log_process("Data saved to CSV file.")

sql_conn = sql.connect(db_name)

log_process("SQL Connection initiated.")

load_to_db(df, sql_conn, table_name)

log_process("Data loaded to Database as table. Running the query")

run_query(f"SELECT * FROM {table_name} WHERE GDP_USD_billions >= 100", sql_conn)

sql_conn.close()

log_process('Process Complete.')
```
