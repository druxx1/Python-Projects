## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados.Bairro.isin(bairros)
dados = dados[selecao]
```

```python
dados.Bairro.drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
grupo_bairro.groups
```

```python
print('Valor medio do aluguel em:')
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data['Valor'].mean()))
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatisticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min': 'Minimo', 'max': 'Maximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_title('Valor Medio do Aluguel por Bairro', {'fontsize': 21})
fig.set_ylabel('Valor Medio', {'fontsize': 14})
fig.set_xlabel('Bairros', {'fontsize': 14})
```

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
```

```python
alunos
```

```python
sexo = alunos.groupby('Sexo')
sexo = pd.DataFrame(sexo['Notas'].mean().round(2))
sexo.columns = ['Notas Medias']
sexo
```

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto')
produtos.describe().round(2)
```

```python
estatisticas = ['mean', 'std', 'min', 'max']
nome = {'mean': 'Media', 'std': 'Desvio Padrao', 'min': 'Minimo', 'max': 'Maximo'}
produtos['Preço'].aggregate(estatisticas).rename(columns = nome).round(2)
```
