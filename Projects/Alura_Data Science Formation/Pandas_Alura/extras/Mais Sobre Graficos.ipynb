```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1) # shift tab
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados['Valor'], dados['Area'])
g1.set_title('Valor x Area')

g2.hist(dados['Valor'])
g2.set_title('Histograma')

dados_g3 = dados['Valor'].sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Medio por Tipo')
```

```python
area.savefig('grafico.png', dpi = 300, bbox_inches = 'tight')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
```

```python
data.head(15)
```

```python
fig = plt.figure()
```

```python
G1 = fig.add_subplot(1, 2, 1)
G1.set_title('Total de Imoveis por Tipo Agregado')
grupo1 = data.groupby('Tipo Agregado')['Valor']
label1 = grupo1.count().index
valores1 = grupo1.count().values
G1.pie(valores1, labels = label1, autopct='%1.1f%%')

G2 = fig.add_subplot(1, 2, 2)
G2.set_title('Total de Imoveis por TIpo')
grupo2 = data.groupby('Tipo')['Valor']
label2 = grupo2.count().index
valores2= grupo2.count().values
G2.pie(valores2, labels = label2, autopct='%1.1f%%')
```
