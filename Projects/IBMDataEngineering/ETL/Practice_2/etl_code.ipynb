```python
import pandas as pd
import glob
import xml.etree.ElementTree as ET
from datetime import datetime
```

## Extract

```python
# Criando funcoes para extrair os dados para cada tipo de arquivo especifico

log_file = "log_file.txt"
target_file = "transformed_data.csv"

def extract_from_csv(file_to_process):
    df = pd.read_csv(file_to_process)
    return df

def extract_from_json(file_to_process):
    df = pd.read_json(file_to_process, lines = True)
    return df

def extract_from_xml(file_to_process):
    df = pd.DataFrame(columns = ['car_model','year_of_manufacture','price','fuel'])
    tree = ET.parse(file_to_process)
    root = tree.getroot()
    for car in root:
        car_model = car.find('car_model').text
        year = int(car.find('year_of_manufacture').text)
        price = float(car.find('price').text)
        fuel = car.find('fuel').text
    return df
```

```python
# Criando funcao para extrair os dados em definitivo e colocá-los em um DataFrame

def extract():
    
    # lista vazia para receber os dados
    lst = list()
    
    # Processando todos os arquivos CSV, exceto o 'target_file'
    for csv in glob.glob('*.csv'):
        if csv != target_file: # garantindo que o(s) arquivo(s) noa é o 'target_file'
            lst.append(extract_from_csv(csv))
    
    # Processando todos os arquivos JSON
    for json in glob.glob('*.json'):
        lst.append(extract_from_json(json))
        
    # Processando todos os arquivos XML
    for xml in glob.glob('*.xml'):
        lst.append(extract_from_xml(xml))
        
    extracted_data = pd.concat(lst, ignore_index = True)
    return extracted_data
```

```python
# Rodando/Visualizando os dados do processo de extracao antes de transformar
extracted_data = extract()
extracted_data
```

## Transform

```python
# Criando funcao para o processo de transformacao dos dados
def transform(data):
    data['price'] = round(data.price, 2)
    
    return data
```

```python
# Visualizando transformacao dos dados
transformed_data = transform(extracted_data)
transformed_data
```

## Load

```python
# Criando o processo de 'load' dos dados e passando os dados transformados para o 'target_file'
def load_data(target_file, transformed_data):
    transformed_data.to_csv(target_file)
```

```python
# Criando a funcao para escrever o 'log' do processo ETL que acabamos de criar
def log_progress(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Ano-Mes-Dia-Hora-Minutos-Segundos
    now = datetime.now() # timestamp atual
    timestamp = now.strftime(timestamp_format)
    
    with open(log_file,"a") as file:
        file.write(timestamp + "," + message + "\n")
```

```python
# Inicio do processo ETL
log_progress("ETL Job Started")
 
# Inicio do processo de extracao
log_progress("Extract phase Started")
#extracted_data = extract() 
 
# Processo de extracao finalizado
log_progress("Extract phase Ended")
 
# Inicio do processo de transformacao
log_progress("Transform phase Started")
 
# Processo de transformacao finalizado
log_progress("Transform phase Ended")
 
# Inicio do processo de 'load'
log_progress("Load phase Started")
load_data(target_file,transformed_data)
 
# Processo de 'load' finalizado
log_progress("Load phase Ended")
 
# Processo de ETL finalizado
log_progress("ETL Job Ended")

# Visualizando os dados após o ETL + Log
display(pd.read_csv('transformed_data.csv', index_col=0))
with open(log_file,"r") as file:
    print(file.read())
```
